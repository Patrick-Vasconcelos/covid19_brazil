```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv("owid-covid-data.csv")
```

```python
df
```

```python
df.info()
```

```python
df_brasil = df.query('location == "Brazil"')
```

```python
df_brasil
```

```python
df_brasil_ = df_brasil[['date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_tests',
                        'new_tests', 'total_vaccinations','new_vaccinations', 'people_vaccinated',
                        'population']]
df_brasil_
```

```python
pd.to_datetime(df_brasil_['date'])
```

```python
df_brasil_['date'] = pd.to_datetime(df_brasil_['date'], format='%Y-%m-%d')
```

```python
df_brasil_
```

```python

df_brasil_.plot(x = 'date', y = 'new_cases')
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(data=df_brasil_, x = 'date', y = 'new_cases')
plt.ylabel("Novos casos")
plt.xlabel("")
plt.title("Novos casos no brasil")
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(data=df_brasil_, x = 'date', y = 'total_vaccinations')
plt.ylabel("Vacinados")
plt.xlabel("")
plt.title("Pessoas Vacinadas no brasil")
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(data=df_brasil_, x = 'date', y = 'total_deaths')
plt.ylabel("Mortes")
plt.xlabel("")
plt.title("Mortes no brasil")
plt.show()
```

```python
df_brasil_janeiro = df_brasil_.query("date > '2021-11-1'")
```

```python
fig = plt.figure(figsize = (20,10))
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(datetime.datetime(2021,11,1),datetime.datetime(2022,1,31))
ax.plot(df_brasil_janeiro['date'], df_brasil_janeiro['new_deaths'])
ax.grid(True)
plt.ylabel("Novas Mortes", fontsize= 15)
plt.xlabel("")
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.title("Novas Mortes no Mês de Janeiro no Brasil", fontsize = 20)
ax.axhline(df_brasil_janeiro['new_deaths'].max(), color = 'k', linestyle = '--')
plt.show()
```
